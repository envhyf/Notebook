
__IE Calibration__

__26.05.2021__ 

__Ionization efficiency__, a dimensionless quantity equaling the number of ions detected per molecule for specific species. It is very important for AMS to transform signals in unit of HZ to mass conc: 
  

$$C = \frac{10^{12} \cdot MW}{IE\cdot Q\cdot N_{A}}\cdot I ^{[1]}$$   

* Q, flow rate into the instrument, in $cm^{3} s^{-1}$.    
* Mw, molecular weight of the species, in $g mol^{-1}$.    
* NA, Avogadro’s number.
* I, ion rate, count per second, in $HZ$.     



---
Today, we measured several NH4NO3 solutions via both AMS and SMPS, which could reproduce the observed mass concentrations and number concentrations, respectively.

We will treat number concentrations from SMPS as true vaules, and then correct the AMS IE values by compare the results from two instruments.
     

---

SMPS Number concentration to mass concentration   

$$C_{m} = S\cdot \rho \cdot \frac{4}{3} \pi (D/2)^{3} \cdot C_{n}$$  


* S, shape factor, parameter relating aerodynamic and geometric diameter 
* $\rho$, mass density
* $D$, aerodynamic diameter of the particle 
* $C_{n}$, number concentration, in $p/cc$  



---
In our case for $NH_{4}NO_{3}$:  

$S = 0.8^{[2]}$, $\rho=1.725\ g\ cm^{-3}$, $D =300\ nm$    

Thus,

$C_{m} = 0.8\cdot(10^{12})\cdot 1.725\cdot4/3\cdot 3.14\cdot (300/2 \cdot 10^{-9})^{3} \cdot 10^{6} \cdot C_{n}$ = 0.0195 $C_{n}$      


Also note: AMS observed nitrate concentrations were calculated by:

$$Nitrate_{Pika} = HRNO3*(M_{NO_3}+M_{NH_4})/M_{NO_3} = 1.299\cdot HRNO3$$
